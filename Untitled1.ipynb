```python
# Function to identify the sub-ingrediants in the ingrediant texts by counting their paratheses and put the result into list
def find_inside(text,i):
  if text.find('(',i) is not None:
      f_open = text.find('(',i)+1
  else: f_open = 0
  if text.find(')',f_open) is not None:
     close = text.find(')',f_open)
  else: close = 0
  while text[f_open:close].count('(') != text[f_open:close].count(')'):
    close = text.find(')',close+1)
  else: f_close = close
  return f_open,f_close

def inside_par (num_embedded,dat):
    text = dat.ingredients_text
    i = 0
    count = 0
    for j in range(num_embedded):
        open,close = find_inside(text,i)
        if open !=0 and close !=0:
          count += dat.ingredients_text[open:close].count(',')
          count_row = dat.ingredients_text[0:close].count(',') - count
          if count_row >=0:
            dat['string'+str(j)] = dat.ingredients_text[open:close]
            dat['ing_string'+str(j)] = count_row
        else: 
            break
        i = close+1
    return dat
```
