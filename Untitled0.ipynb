```python
def find_inside(text,i):
  if text.find('(',i) is not None:
      f_open = text.find('(',i)+1
  else: f_open = 0
  if text.find(')',f_open) is not None:
     close = text.find(')',f_open)
  else: close = 0
  while text[f_open:close].count('(') != text[f_open:close].count(')'):
    close = text.find(')',close+1)
  else: f_close = close
  return f_open,f_close
```
